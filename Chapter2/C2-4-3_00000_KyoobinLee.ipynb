```python
import numpy as np
from matplotlib import pyplot as plt
```

Generate dataset

y = 1*x + noise

```python
N = 100
x = np.linspace(0.0, 1.0, N)
y_gt = x + np.random.normal(0,0.1,N)
plt.plot(x, y_gt, '.')
plt.show()
```

Draw the loss surface with all dataset

```python
for w in np.linspace(0.0, 2.0, N):
    y = w*x
    
    L = np.sum(np.power(y_gt-y, 2))
    dL_dy = sum(-2*(y_gt-y))
    plt.plot(w,L,'.k')
plt.show()
```

Draw the loss surface with mini-batch with n_batch=5


```python
n_batch = 5
for i in range(3):
    randperm = np.random.permutation(N)
    randperm = randperm[:n_batch]
    x2 = x[randperm]
    print(x2)
    y_gt2 = y_gt[randperm]
    
    color = np.random.uniform(0, 0.5, 3)
    for w in np.linspace(0.0, 2.0, N):
        y = w*x2
        
        L = np.sum(np.power(y_gt2-y, 2))
        dL_dy = sum(-2*(y_gt2-y))
        plt.plot(w,L,'.', color=color)
plt.show()
    
```
